```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[1:6])
```

## Numpy Assignment

Create a 1D NumPy array A containing integers from 0 to 23.

Reshape A into a 3D array B of shape (2, 3, 4).

From B, extract all elements that belong to:

the second 3×4 block, and

odd column indices only
Store the result in a new array C.

Flatten C into a 1D array D.

Using np.nditer():

Replace every element in D that is divisible by 3 with –1

Replace every element divisible by 5 with –2

Reshape D into a 2D array E with 2 columns (rows inferred automatically).

Delete:

the first row

the last column

Append a column of ones to the right side of the resulting array.

Print the final array and its shape.

```python
import numpy as np
# Import NumPy library and alias it as np for convenience

# Step 1: Create a 1D array containing numbers from 0 to 23
A = np.arange(24)
# np.arange(24) generates values [0, 1, 2, ..., 23]

print("A:\n", A)
# Print the original 1D array

# Step 2: Reshape the 1D array into a 3D array of shape (2, 3, 4)
B = A.reshape(2, 3, 4)
# 2 blocks, each containing a 3x4 matrix

print("\nB:\n", B)
# Print the 3D array

# Step 3: Extract elements from:
# - second block (index 1)
# - all rows (:)
# - only odd column indices (1, 3)
C = B[1, :, 1::2]
# 1::2 means start from index 1 and take every second column

print("\nC:\n", C)
# Print the extracted sub-array

# Step 4: Flatten C into a 1D array
D = C.flatten()
# Converts multi-dimensional array into a 1D array

print("\nD:\n", D)
# Print flattened array

# Step 5: Iterate over D and modify elements in-place
for x in np.nditer(D, op_flags=['readwrite']):
    # np.nditer allows element-wise iteration
    # readwrite flag allows modifying values

    if x % 3 == 0:
        # Check if element is divisible by 3
        x[...] = -1
        # Replace value with -1

    elif x % 5 == 0:
        # Check if element is divisible by 5
        x[...] = -2
        # Replace value with -2

print("\nD after modification:\n", D)
# Print modified array

# Step 6: Reshape D into a 2D array with 2 columns
E = D.reshape(-1, 2)
# -1 automatically calculates the required number of rows

print("\nE:\n", E)
# Print reshaped array

# Step 7: Delete the first row (row index 0)
E = np.delete(E, 0, axis=0)
# axis=0 refers to rows

# Delete the last column (column index -1)
E = np.delete(E, -1, axis=1)
# axis=1 refers to columns

print("\nE after deletion:\n", E)
# Print array after deletions

# Step 8: Create a column of ones with the same number of rows as E
ones_col = np.ones((E.shape[0], 1))
# E.shape[0] gives number of rows

# Append the column of ones to the right of E
final = np.hstack((E, ones_col))
# hstack horizontally stacks arrays column-wise

print("\nFinal Array:\n", final)
# Print the final array

print("\nFinal Shape:", final.shape)
# Print shape of final array
```

# Scipy

SciPy is a scientific computation library that uses NumPy underneath.
SciPy stands for Scientific Python.
It provides more utility functions for optimization, stats and signal processing.

```python
from scipy import constants

print(constants.liter)
```

**Optimizers in SciPy:** Optimizers are a set of procedures defined in SciPy that either find the minimum value of a function, or the root of an equation.

```python
from scipy.optimize import root
from math import cos

def eqn(x):
  return x + np.cos(x)

myroot = root(eqn, 0)

print(myroot.x[0])
```

Here’s what happens during the iterative process:

At (x = 0), (f(x) = 1), so the guess is adjusted to reduce (f(x)).

At the next step, say (x = -0.5): [ f(-0.5) = -0.5 + \cos(-0.5) \approx -0.5 + 0.877 = 0.377 ] Since (f(x) \neq 0), the algorithm adjusts (x) further.

At (x = -0.7): [ f(-0.7) = -0.7 + \cos(-0.7) \approx -0.7 + 0.764 = 0.064 ] The value is getting closer to 0.

At (x = -0.739): [ f(-0.739) = -0.739 + \cos(-0.739) \approx -0.739 + 0.739 = 0 ] The root is found.

```python
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

points = np.array([
  [2, 4],
  [3, 4],
  [3, 0],
  [2, 2],
  [4, 1]
])

simplices = Delaunay(points).simplices

plt.triplot(points[:, 0], points[:, 1], simplices)
plt.scatter(points[:, 0], points[:, 1], color='r')

plt.show()
```

`Delaunay` from `scipy.spatial` is used for performing Delaunay triangulation, which is a way to divide a set of points into triangles while respecting certain geometric properties.

 `matplotlib.pyplot` is used for plotting and visualizing the triangulation and the data points.

 `Simplices` contains an array of indices that represent the vertices of each triangle formed by the triangulation.

 `points[:,0]` extracts all the x-coordinates of the points.

 `points[:,1]` extracts all the y-coordinates of the points.

 `plt.scatter` plots the original points as red dots on the graph for reference.

**Interpolation** is a method for generating points between given points.

Interpolation has many usage, in Machine Learning we often deal with missing data

Question

1. You are given experimental data obtained from a scientific simulation involving geometry and numerical modeling.

The volume of a container is given as 1 liter.
Convert this volume into cubic meters using SciPy’s physical constants.

2. The system behavior is governed by the nonlinear equation:

f(x)=x+cos(x)

Find a real root of this equation using a numerical method, starting from an initial guess of 0.

3. Sensor readings are available at discrete points:

𝑥 = [0,1,2,...,9],   𝑦 = 2𝑥 + 1

Construct a continuous approximation of this data using interpolation and estimate the values of the function for:

x=2.1,2.2,...,2.9

4. The spatial coordinates of measurement points in a 2D plane are:

(2,4),(3,4),(3,0),(2,2),(4,1)

Perform Delaunay triangulation on these points and visualize the resulting mesh along with the original points.



```python
# Import NumPy for numerical operations and array handling
import numpy as np

# Import matplotlib for plotting graphs and visualizations
import matplotlib.pyplot as plt

# Import cosine function from math module
from math import cos

# Import physical constants from SciPy
from scipy import constants

# Import root-finding function from SciPy optimization module
from scipy.optimize import root

# Import 1D interpolation function from SciPy
from scipy.interpolate import interp1d

# Import Delaunay triangulation class from SciPy spatial module
from scipy.spatial import Delaunay

# -------------------------------------------------
# PART 1: Conversion of volume using SciPy constants

# Convert 1 liter into cubic meters using SciPy constant
volume_m3 = constants.liter

# Print the converted volume
print("Volume in cubic meters:", volume_m3)

# -------------------------------------------------
# PART 2: Root finding for nonlinear equation

# Define the mathematical function f(x) = x + cos(x)
def eqn(x):
    # Return the value of the equation at x
    return x + cos(x)

# Use SciPy's root function to find a root of eqn starting from initial guess 0
solution = root(eqn, 0)

# Extract and print the root value rounded to 4 decimal places
print("Root of equation:", round(solution.x[0], 4))

# -------------------------------------------------
# PART 3: Interpolation of discrete data points

# Create an array of x-values from 0 to 9
xs = np.arange(10)

# Compute corresponding y-values using y = 2x + 1
ys = 2 * xs + 1

# Create a 1D interpolation function based on xs and ys
interp_func = interp1d(xs, ys)

# Generate new x-values between 2.1 and 2.9 with step size 0.1
new_x = np.arange(2.1, 3, 0.1)

# Use the interpolation function to estimate y-values at new_x
new_y = interp_func(new_x)

# Print the interpolated values
print("Interpolated values:", new_y)

# -------------------------------------------------
# PART 4: Delaunay triangulation and visualization

# Define a set of 2D points as a NumPy array
points = np.array([
    [2, 4],
    [3, 4],
    [3, 0],
    [2, 2],
    [4, 1]
])

# Perform Delaunay triangulation on the given points
triangles = Delaunay(points).simplices

# Plot the triangulated mesh using the triangulation indices
plt.triplot(points[:, 0], points[:, 1], triangles)

# Plot the original points as red dots
plt.scatter(points[:, 0], points[:, 1], color='red')

# Add a title to the plot
plt.title("Delaunay Triangulation")

# Display the plot on screen
plt.show()
```
